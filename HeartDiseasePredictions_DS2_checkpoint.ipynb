Outliers Detection

```python
from scipy import stats
dt_heart_out = df_heart[['Age','RestingBP','Cholesterol','MaxHR']]
# calculating the Z-score
z = np.abs(stats.zscore(dt_heart_out))
print(z)
```

```python
threshold = 3
print(np.where(z > 3))
```

```python
#filter the outliers and removal
df_new_heart = df_heart[(z < 3).all(axis=1)]
df_new_heart.shape
```

12 outliers are removed

Data Transfomartion

```python
#Data Transformation for categorical value
#One Hot Encoding
df_new_columns = pd.get_dummies(data = df_new_heart, columns = ['Sex','ChestPainType','RestingECG','ST_Slope','ExerciseAngina'])
df_new_columns.columns
```

```python
#After converting the categorical variable .print the frist five rows
df_new_columns.head()
```

```python
#Data Transformation for numerical value
col=['Age','RestingBP','Cholesterol','FastingBS','MaxHR','Oldpeak']
from sklearn.preprocessing import MinMaxScaler
MMS=MinMaxScaler()
df_new_columns[col]= MMS.fit_transform(df_new_columns[col])
df_new_columns.head()
```

```python
#Dependent and Independent
X = df_new_columns.drop(columns = ['HeartDisease'])
Y = df_new_columns['HeartDisease']
```

```python
#Data splitting
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.50,random_state= 42)
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

Model Fitting

Logistic Regression

```python
#Logisitic Algorithm
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X_train,Y_train)
LR.score(X_train,Y_train).round(2)
```

```python
#checking the precision,recall,f1-score,accuracy_score.
from sklearn.metrics import classification_report,accuracy_score
Y_pred_train = LR.predict(X_train)
Y_pred_test = LR.predict(X_test)
print(classification_report(Y_pred_test,Y_test))
```

```python
#generating the roc_auc curve
from sklearn.metrics import roc_curve,roc_auc_score
def generate_auc_roc_curve(clf, X_test):
    y_pred_proba = LR.predict_proba(X_test) [:, 1]
    fpr, tpr, thresholds = roc_curve(Y_test, y_pred_proba)
    auc = roc_auc_score(Y_test, y_pred_proba)
    plt.plot(fpr, tpr, label="AUC ROC Curve with Area Under the curve "+str(auc))
    plt.legend(loc=4)
    plt.show()
```

```python
generate_auc_roc_curve(LR,X_test)
```

```python
#model validation for Logistic Regression
from sklearn.model_selection import StratifiedKFold,cross_val_score
K = StratifiedKFold(n_splits = 10)
result_LR = cross_val_score(LR,X,Y,cv = K)
print(result_LR)
```

```python
print("StratifiedKFold for logistic regression is",np.mean(result_LR).round(2))
```
